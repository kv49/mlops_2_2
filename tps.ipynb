```python
import os
from pathlib import Path
import pandas as pd # Библиотека Pandas для работы с табличными данным
#from google.colab import files  # чтобы загружать файлы в облако через проводник
import numpy as np # библиотека Numpy для операций линейной алгебры и прочего

from sklearn.model_selection import train_test_split
```

```python
path = Path('.')

submission = pd.read_csv(path / 'sample_submission.csv', index_col='id')
labels = pd.read_csv(path / 'train_labels.csv', index_col='id')

# the ids of the submission rows (useful later)
sub_ids = submission.index

# the ids of the labeled rows (useful later)
gt_ids = labels.index 

# list of files in the submission folder
subs = sorted(os.listdir(path / 'submission_files'))
```

```python
X_all = np.array([])
for i in np.arange(len(subs)):
    tmp = pd.read_csv(path / 'submission_files' / subs[i], index_col='id')
    X_all = np.append(X_all, np.array(tmp['pred']), axis=0)
```

```python
TEST_SIZE = float(0.0)
SEED = 42
EPOCH = 9
```

```python
if TEST_SIZE > 0:
    train_ids, test_ids, train_labels, test_labels = \
        train_test_split(labels1['id'], labels1['label'], test_size=TEST_SIZE, random_state=SEED, stratify=labels1['label'])
    df_train = X_all_df[X_all_df['id'].isin(train_ids)]
    df_test = X_all_df[X_all_df['id'].isin(test_ids)]
else:
    train_ids, test_ids, train_labels, test_labels = labels1['id'], pd.Series(), labels1['label'], pd.Series()
    df_train = X_all_df[0:20000]
    df_test = X_all_df[0:0]
```

```python
df_train
```

```python
train_labels
```

```python
x_train = df_train.loc[:, '0':].to_numpy()
y_train = labels1[labels1['id'].isin(train_ids)]['label'].to_numpy()

x_test = df_test.loc[:, '0':].to_numpy()
y_test = labels1[labels1['id'].isin(test_ids)]['label'].to_numpy()
```

```python
%%time
from sklearn.neural_network import MLPClassifier  # многослойный персептрон (просто много полносвязных слоев)
clf = MLPClassifier(hidden_layer_sizes=(5000,), max_iter=EPOCH, random_state=SEED, verbose=True)
clf.fit(x_train, y_train)
```

```python
test_pred = clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, test_pred)
```

```python
test_pred
```

```python
from sklearn.metrics import log_loss
score = log_loss(y_test, test_pred)
score
```

```python
from datetime import datetime
```

```python
mlp_stat_df = pd.DataFrame(columns=['iteration', 'duration', 'train_loss', 'test_accuracy_score', 'test_log_loss'])
for i in range(1,25):
    start_date = datetime.now()
    clf = MLPClassifier(hidden_layer_sizes=(5000,), max_iter=i, random_state=SEED)
    clf.fit(x_train, y_train)
    iter_res = [i, datetime.now() - start_date,
                clf.loss_,
                accuracy_score(y_test, clf.predict(x_test)),
                log_loss(y_test, clf.predict_proba(x_test))]
    mlp_stat_df.loc[len(mlp_stat_df)] = iter_res
    print(iter_res)
```
